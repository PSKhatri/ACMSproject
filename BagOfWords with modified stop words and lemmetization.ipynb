```python
#loads the csv file into data frame
import pandas as pd

df = pd.read_csv('/home/pooja/projectACMS/train_lyrics_1511.csv')

df.head()
```

```python
happyLyrics=[]
sadLyrics=[]
for row_id in df.index:
    print(row_id)
    if df.loc[row_id]['mood'] == 'happy':
        happyLyrics.append(df.loc[row_id]['lyrics'])
    else:
        sadLyrics.append(df.loc[row_id]['lyrics'])
    
```

```python
happy_Lyrics=[]
for w in happyLyrics:
    happy_Lyrics.append(w.lower())
#print(happy_Lyrics)   
sad_Lyrics=[]
for w in sadLyrics:
    sad_Lyrics.append(w.lower())
#print(sad_Lyrics) 
```

```python
import nltk
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
happyWords=(word_tokenize( ' '.join(happy_Lyrics)))
sadWords=(word_tokenize( ' '.join(sad_Lyrics)))
#print(happyWords)
```

```python
with open('/home/pooja/projectACMS/stopwords1.txt', 'r') as infile:
    stop_words = infile.read().splitlines()
#print('stop words %s ' %stop_words)
```

```python
happyWords1=[]
for w in happyWords:
    if w not in stop_words:
        happyWords1.append(w)
#print(happyWords1) 

sadWords1=[]
for w in sadWords:
    if w not in stop_words:
        sadWords1.append(w)
#print(sadWords1) 
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
happyWords2=[]
#stemmer = PorterStemmer()
lemmatiser = WordNetLemmatizer()
for w in happyWords1:
    happyWords2.append(lemmatiser.lemmatize(w, pos="v"))
#print(happyWords2) 

sadWords2=[]
#stemmer = PorterStemmer()
lemmatiser = WordNetLemmatizer()
for w in sadWords1:
    sadWords2.append(lemmatiser.lemmatize(w, pos="v"))
#print(sadWords2) 
```

```python
happyWords3=[]
for w in happyWords2:
    if w not in stop_words:
        happyWords3.append(w)
#print(happyWords3) 

sadWords3=[]
for w in sadWords2:
    if w not in stop_words:
        sadWords3.append(w)
#print(sadWords3) 
```

```python
import re
no_punct_happy = [s for s in happyWords3 if re.match('^[a-zA-Z]+$', s) is not None]
#print(no_punct_happy)

no_punct_sad = [s for s in sadWords3 if re.match('^[a-zA-Z]+$', s) is not None]
#print(no_punct_sad)
```

```python
print(len(no_punct_happy))
print(len(no_punct_sad))
```

```python
from collections import Counter
```

```python
happy3=Counter(no_punct_happy)
print(len(happy3))
print(happy3)
```

```python
sad3=Counter(no_punct_sad)
print(len(sad3))
print(sad3)
```

```python
count=0
for row_id in df1.index:
    text=df1.loc[row_id]['lyrics']
    text1=(word_tokenize(text))
    text2=[]
    for w in text1:
        text2.append(w.lower())
    text3=[]
    for w in text2:
        if w not in stop_words:
            text3.append(w)
    text4=[]
    for w in text3:
        text4.append(lemmatiser.lemmatize(w, pos="v"))
    text5 = [s for s in text4 if re.match('^[a-zA-Z]+$', s) is not None] 
    sum_happy=0
    sum_sad=0
    for i in text5:
        if i in happy3 and i in sad3:
            sum_happy=sum_happy+happy3[i]/(happy3[i]+sad3[i])
            sum_sad=sum_sad+sad3[i]/(happy3[i]+sad3[i])
        elif i in happy3 and i not in sad3:
            sum_happy=sum_happy+1
        elif i not in happy3 and i in sad3:
            sum_sad=sum_sad+1
    
    if sum_happy>sum_sad:
        pred_mood='happy'
    elif sum_sad>sum_happy:
        pred_mood='sad'
        
    if pred_mood==df1.loc[row_id]['mood']:
        count=count+1
    print(row_id,   pred_mood,  df1.loc[row_id]['mood'], count)
        
        
print(count)  
print("acc",count/200)
        
```
